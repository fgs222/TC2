Del circuito obtenemos las siguientes matrices:

$$
\begin{pmatrix} V_1\\0 \end{pmatrix} = 
\underbrace{
    \begin{pmatrix} Z_C+Z_D & -Z_D\\-Z_D & Z_D+Z_R+Z_L \end{pmatrix}
    }_{Z(s)} . 
\begin{pmatrix} I_1\\I_2 \end{pmatrix} 
$$

Resolviendo por metodo de mallas:

$$
|Z(s)| = 
\begin{vmatrix}
    Z_C+Z_D & -Z_D\\-Z_D & Z_D+Z_R+Z_L
\end{vmatrix}
$$

$$
Z(s) = Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_D^2+Z_DZ_R+Z_DZ_L-Z_D^2
$$

$$
Z(s) = Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_DZ_R+Z_DZ_L
$$

Siendo:

$$
I_2 = \frac{\begin{vmatrix}
        Z_C+Z_D & V_1\\-Z_D & 0
    \end{vmatrix}}
    {\begin{vmatrix}
        Z_C+Z_D & -Z_D\\-Z_D & Z_D+Z_R+Z_L 
    \end{vmatrix}} =
    \frac{V_1 Z_D}
    {Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_DZ_R+Z_DZ_L}
$$

$$
I_2 = ~V_1~~\frac{Z_D}{Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_DZ_R+Z_DZ_L}~~(1)
$$

Siendo:
$$
V_2 = Z_L.I_2~~(2)
$$

Remplazando (1) en (2):

$$
V_2 = Z_L~~V_1~~\frac{Z_D}{Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_DZ_R+Z_DZ_L}
= V_1 \frac{Z_DZ_L}{Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_DZ_R+Z_DZ_L}
$$

$$
\frac{V_2}{V_1} = H(s) = \frac{Z_DZ_L}{Z_CZ_D+Z_CZ_R+Z_CZ_L+Z_DZ_R+Z_DZ_L}
$$

Siendo:

$$
Z_C= \frac{1}{sC}~~~~~~Z_D = \frac{1}{s^2D}~~~~~~Z_R = R~~~~~~Z_L = sL
$$

$$
H(s) = \frac{\frac{1}{s^2D}sL}{\frac{1}{sC}\frac{1}{s^2D} + \frac{1}{sC}R + \frac{1}{sC}sL + \frac{1}{s^2D}R + \frac{1}{s^2D}sL}
= \frac{\frac{L}{D}}{s(\frac{1}{sC}\frac{1}{s^2D} + \frac{1}{sC}R + \frac{1}{sC}sL + \frac{1}{s^2D}R + \frac{1}{s^2D}sL)}
$$

$$
H(s) = \frac{\frac{L}{D}}{s(\frac{1}{s^3DC} + \frac{R}{sC} + \frac{L}{C} + \frac{R}{s^2D} + \frac{L}{sD})}
= \frac{\frac{L}{D}}{\frac{1}{s^2DC} + \frac{R}{C} + s\frac{L}{C} + \frac{R}{sD} + \frac{L}{D}}\frac{s^2}{s^2}
$$

$$
H(s) = \frac{s^2\frac{L}{D}}{\frac{1}{DC} + s^2\frac{R}{C} + s^3\frac{L}{C} + s\frac{R}{D} + s^2 \frac{L}{D}}
= \frac{s^2L}{D(s^3\frac{L}{C} + s^2\frac{R}{C} + s^2 \frac{L}{D} + s\frac{R}{D} + \frac{1}{DC})}
$$

$$
H(s) = \frac{s^2L}{s^3\frac{DL}{C} + s^2\frac{DR}{C} + s^2L + sR + \frac{1}{C}}\frac{C}{C}
= \frac{s^2LC}{s^3DL + s^2DR + s^2LC + sRC + 1}
$$

$$
H(S) = \frac{s^2LC}{s^3DL + s^2(DR+LC) + sRC + 1}
$$
