$$
\frac{V_i-V_x}{Z_5}=\frac{V_x-V_2}{Z_6}~~(1)~~~~~~\frac{V_2}{Z_2}+\frac{V_o}{Z_b}=0~~(2)~~~~~~\frac{V_1}{Z_1}+\frac{V_2}{Z_a}=0~~(3)~~~~~~\frac{V_o-V_x}{Z_3}=\frac{V_x-V_1}{Z_4}~~(4)
$$

De (2):

$$
V_2 = -V_o\frac{Z_2}{Z_b}~~(5)
$$

De (3):

$$
V_1 = -V_2\frac{Z_1}{Z_a}~~(6)
$$

Remplazando (5) en (6):

$$
V_1 = V_o\frac{Z_1Z_2}{Z_aZ_b}~~(7)
$$

De (1):

$$
\frac{V_i}{Z_5}-\frac{V_x}{Z_5} = \frac{V_x}{Z_6}-\frac{V_2}{Z_6}~~\to~~\frac{V_i}{Z_5}+\frac{V_2}{Z_6} = \frac{V_x}{Z_6}+\frac{V_x}{Z_5}~~\to~~\frac{V_i}{Z_5}+\frac{V_2}{Z_6} = V_x\frac{Z_5+Z_6}{Z_5Z_6}
$$

$$
V_x = \frac{Z_5Z_6}{Z_5+Z_6}(\frac{V_i}{Z_5}+\frac{V_2}{Z_6})~~(8)
$$

De (4):

$$
\frac{V_o}{Z_3}-\frac{V_x}{Z_3} = \frac{V_x}{Z_4}-\frac{V_1}{Z_4}~~\to~~\frac{V_o}{Z_3}+\frac{V_1}{Z_4} = \frac{V_x}{Z_3}+\frac{V_x}{Z_4}~~\to~~\frac{V_o}{Z_3}+\frac{V_1}{Z_4} = V_x\frac{Z_3+Z_4}{Z_3Z_4}
$$

Remplazando (7) en V1 y (8) en Vx:

$$
\frac{V_o}{Z_3}+V_o\frac{Z_1Z_2}{Z_aZ_bZ_4} = \frac{Z_5Z_6}{Z_5+Z_6}\frac{Z_3+Z_4}{Z_3Z_4}(\frac{V_i}{Z_5}-\frac{V_2}{Z_6})
$$

Remplazando (5) en V2:

$$
\frac{V_o}{Z_3}+V_o\frac{Z_1Z_2}{Z_aZ_bZ_4} = \frac{Z_5Z_6(Z_3+Z_4)}{Z_3Z_4(Z_5+Z_6)}(\frac{Vi}{Z_5}-V_o\frac{Z_2}{Z_bZ_6})
$$

$$
\frac{V_o}{Z_3}+V_o\frac{Z_1Z_2}{Z_aZ_bZ_4}+V_o\frac{Z_2Z_5(Z_3+Z_4)}{Z_bZ_3Z_4(Z_5+Z_6)} = V_i\frac{Z_6(Z_3+Z_4)}{Z_3Z_4(Z_5+Z_6)}
$$

$$
V_o\frac{Z_aZ_bZ_4(Z_5+Z_6)+Z_3(Z_5+Z_6)+Z_aZ_2Z_5(Z_3+Z_4)}{Z_aZ_bZ_3Z_4(Z_5+Z_6)} = V_i\frac{Z_6(Z_3+Z_4)}{Z_3Z_4(Z_5+Z_6)}
$$

$$
\frac{V_o}{V_i} = \frac{Z_6(Z_3+Z_4)}{Z_aZ_bZ_4(Z_5+Z_6)+Z_3(Z_5+Z_6)+Z_aZ_2Z_5(Z_3+Z_4)}\frac{Z_aZ_bZ_3Z_4(Z_5+Z_6)}{Z_3Z_4(Z_5+Z_6)}
$$

Se llega a la expresion de C:

$$
\frac{V_o}{V_i} = \frac{Z_aZ_bZ_6(Z_3+Z_4)}{Z_aZ_bZ_4(Z_5+Z_6)+Z_3(Z_5+Z_6)+Z_aZ_2Z_5(Z_3+Z_4)}
$$

Siendo:

$$
Z_a = \frac{1}{sC_2}~~~~Z_b = \frac{1}{sC_3}~~~~Z_1 = R_1~~~~Z_2 = R_2~~~~Z_3 = R_3~~~~Z_4 = R_4~~~~Z_5 = R_5~~~~Z_6 = R_6
$$

Remplazando en la ecuacion:

$$
\frac{V_o}{V_i} = \frac{\frac{1}{sC_2}\frac{1}{sC_3}R_6(R_3+R_4)}{\frac{1}{sC_2}\frac{1}{sC_3}R_4(R_5+R_6)+R_3(R_5+R_6)+\frac{1}{sC_2}R_2R_5(R_3+R_4)}
$$

$$
\frac{V_o}{V_i} = \frac{\frac{1}{s^2C_2C_3}R_6(R_3+R_4)}{\frac{1}{s^2C_2C_3}R_4(R_5+R_6)+\frac{1}{sC_2}R_2R_5(R_3+R_4)+R_3(R_5+R_6)}\frac{s^2}{s^2}
$$

$$
\frac{V_o}{V_i} = \frac{\frac{R_6(R_3+R_4)}{C_2C_3}}{s^2R_3(R_5+R_6)+s\frac{R_2R_5(R_3+R_4)}{C_2}+\frac{R_4(R_5+R_6)}{C_2C_3}}
$$

$$
\frac{V_o}{V_i} = \frac{1}{R_3(R_5+R_6)}\frac{\frac{R_6(R_3+R_4)}{C_2C_3}}{S^2+s\frac{R_2R_5(R_3+R_4)}{R_3(R_5+R_6)C_2}+\frac{R_4(R_5+R_6)}{R_3(R_5+R_6)C_2C_3}}
$$

$$
\frac{V_o}{V_i} = \frac{\frac{R_6(R_3+R_4)}{R_3(R_5+R_6)C_2C_3}}{S^2+s\frac{R_2R_5(R_3+R_4)}{R_3(R_5+R_6)C_2}+\frac{R_4(R_5+R_6)}{R_3(R_5+R_6)C_2C_3}}\frac{R_4(R_5+R_6)}{R_4(R_5+R_6)}
$$

Se observa que la transferencia corresponde a un filtro pasabajos:

$$
\frac{V_o}{V_i} = \frac{R_6(R_3+R_4)}{R_4(R_5+R_6)}\frac{\frac{R_4(R_5+R_6)}{R_3(R_5+R_6)C_2C_3}}{S^2+s\frac{R_2R_5(R_3+R_4)}{R_3(R_5+R_6)C_2}+\frac{R_4(R_5+R_6)}{R_3(R_5+R_6)C_2C_3}}
$$

$$
R=R_1=R_2=R_3=R_4=R_5=R_6= 1k \Omega~~~~~~C=C_2=C-3= 1\mu F
$$
