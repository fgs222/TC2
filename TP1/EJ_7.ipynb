Del circuito podemos obtener las siguientes ecuaciones:

$$
\frac{V_1-V_a}{Z_1} = \frac{V_a-V_2}{Z_2}~~(1)~~~~~~~~~~~~\frac{V_1-V_a}{Z_4} = \frac{V_a}{Z_3}~~(2)
$$

De (2):

$$
V_a~(\frac{1}{Z_3}+\frac{1}{Z_4}) = \frac{V_1}{Z_4}~~\to~~V_a~\frac{Z_3+Z_4}{Z_3Z_4} = \frac{V_1}{Z_4}~~\to~~V_a = V_1\frac{Z_4}{Z_3+Z_4}~~(3)
$$

De (1):

$$
\frac{V_1}{Z_1}-\frac{V_a}{Z_1} = \frac{V_a}{Z_2}-\frac{V_2}{Z_2}~~\to~~\frac{V_1}{Z_1}-V_a(\frac{1}{Z_1}+\frac{1}{Z_2}) = \frac{-V_2}{Z_2}~~\to~~\frac{V_1}{Z_1}-V_a\frac{Z_1+Z_2}{Z_1Z_2} = \frac{-V_2}{Z_2}
$$

Remplazando (3) en Va:

$$
\frac{V_1}{Z_1}-V_1\frac{Z_4}{Z_3+Z_4}\frac{Z_1+Z_2}{Z_1Z_2} = \frac{-V_2}{Z_2}~~\to~~V_1 \frac{Z_2(Z_3+Z_4)-Z_4(Z_1Z_2)}{Z_1Z_2(Z_3+Z_4)} = \frac{-V_2}{Z_2}
$$

$$
\frac{V_2}{V_1} = -\frac{Z_2(Z_3+Z_4)-Z_4(Z_1Z_2)}{Z_1(Z_3+Z_4)}~~\to~~\frac{V_2}{V_1} = \frac{Z_4(Z_1Z_2)-Z_2(Z_3+Z_4)}{Z_1(Z_3+Z_4)}
$$

$$
\frac{V_2}{V_1} = \frac{Z_1Z_4+Z_2Z_4-Z_2Z_3-Z_2Z_4}{Z_1(Z_3+Z_4)}~~\to~~\frac{V_2}{V_1} = \frac{Z_1Z_4-Z_2Z_3}{Z_1Z_3+Z_1Z_4}
$$

Para el primer circuito:

$$
Z_1 = R_1~~~~~~Z_2 = R_2~~~~~~Z_3 = R_3~~~~~~Z_4 = \frac{1}{sC}~~~~~~~R_1 = R_2
$$

$$
\frac{V_1}{V_2} = \frac{R_1\frac{1}{sC}-R_2R_3}{R_1\frac{1}{sC}+R_1R_3}~~\to~~\frac{V_1}{V_2} = \frac{R_1\frac{1}{sC}-R_1R_3}{R_1\frac{1}{sC}+R_1R_3}~~\to~~\frac{V_1}{V_2} = \frac{R_1}{R_1}\frac{\frac{1}{sC}-R_3}{\frac{1}{sC}+R_3}
$$

$$
\frac{V_1}{V_2} = \frac{\frac{1}{sC}-R_3}{\frac{1}{sC}+R_3}\frac{sC}{sC}~~\to~~\frac{V_1}{V_2} = \frac{1-sR_3C}{1+sR_3C}
$$

```python
from control import pzmap, TransferFunction, bode

R = 1e3
C = 1e-6

num = [-R*C, 1]
den = [R*C, 1]

tf1 = TransferFunction(num,den)
tf1
```

```python
pzmap1 = pzmap(tf1)
```

```python
bode1 = bode(tf1)
```

Para el segundo circuito:

$$
Z_1 = R_a~~~~~~Z_2 = R_b~~~~~~Z_3 = \frac{sRC+1}{sC}~~~~~~Z_4 = \frac{R}{sRC+1}~~~~~~~R_a = 5R_b
$$

$$
\frac{V_2}{V_1} = \frac{R_a\frac{R}{sRC+1}-R_b\frac{sRC+1}{sC}}{R_a\frac{R}{sRC+1}+R_a\frac{sRC+1}{sC}}~~\to~~\frac{V_2}{V_1} = \frac{R_a\frac{R}{sRC+1}-\frac{1}{5}R_a\frac{sRC+1}{sC}}{R_a\frac{R}{sRC+1}+R_a\frac{sRC+1}{sC}}
$$

$$
\frac{V_2}{V_1} = \frac{R_a}{R_a}\frac{\frac{R}{sRC+1}-\frac{sRC+1}{s5C}}{\frac{R}{sRC+1}+\frac{sRC+1}{sC}}~~\to~~\frac{V_2}{V_1} = \frac{\frac{R}{sRC+1}-\frac{sRC+1}{s5C}}{\frac{R}{sRC+1}+\frac{sRC+1}{sC}}
$$

$$
\frac{V_2}{V_1} = \frac{\frac{s5RC-(sRC+1)^2}{s5C(sRC+1)}}{\frac{sRC+(sRC+1)^2}{sC(sRC+1)}}~~\to~~\frac{V_2}{V_1} = \frac{1}{5}\frac{s5RC-(sRC+1)^2}{sRC+(sRC+1)^2}
$$

$$
\frac{V_1}{V_2} = \frac{1}{5}\frac{s5RC-s^2R^2C^2-s2RC-1}{sRC+s^2R^2C^2+s2RC+1}~~\to~~\frac{V_1}{V_2} = \frac{1}{5}\frac{-s^2R^2C^2+s3RC-1}{s^2R^2C^2+s3RC+1}
$$

$$
\frac{V_1}{V_2} = \frac{-1}{5}\frac{s^2R^2C^2-s3RC+1}{s^2R^2C^2+s3RC+1}~~\to~~\frac{V_1}{V_2} = \frac{-1}{5}\frac{R^2C^2}{R^2C^2}\frac{s^2-s\frac{3RC}{R^2C^2}+\frac{1}{R^2C^2}}{s^2+s\frac{3RC}{R^2C^2}+\frac{1}{R^2C^2}}
$$

$$
\frac{V_1}{V_2} = \frac{-1}{5}\frac{s^2-s\frac{3}{RC}+\frac{1}{R^2C^2}}{s^2+s\frac{3}{RC}+\frac{1}{R^2C^2}}
$$

```python
R = 1e3
C = 1e-6
K = (-1/5)

num = [1, -3/(R*C), 1/(R*C)**2]
den = [1, 3/(R*C), 1/(R*C)**2]

tf2 = K*TransferFunction(num,den)
tf2
```

```python
pzmap = pzmap(tf2)
```

```python
bode = bode(tf2)
```
